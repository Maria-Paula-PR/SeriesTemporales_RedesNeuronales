1. Análisis de datos


```python
df1 = pd.read_csv('univariate/Hourly-train.csv')
df1.head()
```

```python
mine = ['H309', 'H291', 'H143', 'H366', 'H174']
```

```python
series = df1[df1 ['V1'].isin(mine)]
series

```

```python
# Verificar nulos por columna para las series seleccionadas
nulos_por_columna = series.isnull().sum()
print(nulos_por_columna)
```

En la siguiente celda elimino los valores nulos, de las series que me tocaron, al cabo solo estan al final de la serie. 

```python
#visualizar mis series con nulos
plt.figure(figsize=(15, 15))
for i, series_id in enumerate(mine, 1):
    series_values = series[series['V1'] == series_id].iloc[:, 1:].values.flatten()
    
    # Crear array de índices
    indices = range(len(series_values))
    
    # Separar valores nulos y no nulos
    mask_nulos = pd.isnull(series_values)
    valores_nulos = series_values[mask_nulos]
    valores_no_nulos = series_values[~mask_nulos]
    indices_nulos = [i for i, m in enumerate(mask_nulos) if m]
    indices_no_nulos = [i for i, m in enumerate(mask_nulos) if not m]
    
    plt.subplot(len(mine), 1, i)
    # Graficar valores no nulos en azul
    plt.plot(indices_no_nulos, valores_no_nulos, 'b-', label='Valores')
    # Graficar valores nulos en rojo
    plt.plot(indices_nulos, [0]*len(indices_nulos), 'ro', label='Nulos')
    
    plt.title(f'Serie {series_id}')
    plt.xlabel('Horas')
    plt.ylabel('Valor')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
#Graficar la serie sin nulos
# 1. Obtiene los valores de la serie filtrando por el ID
# 2. Selecciona todas las columnas excepto la primera (V1) que contiene el ID
# 3. Convierte los valores en un array 1D con flatten()
# 4. Elimina los valores NaN para tener solo datos válidos
plt.figure(figsize=(15, 10))
for i, series_id in enumerate(mine, 1):
    series_values = series[series['V1'] == series_id].iloc[:, 1:].values.flatten()
    series_values = series_values[~pd.isnull(series_values)]
    
    plt.subplot(len(mine), 1, i)
    plt.plot(series_values, linewidth=2)
    plt.title(f'Serie {series_id}')
    plt.xlabel('Horas')
    plt.ylabel('Valor')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tick_params(labelsize=8)

plt.tight_layout()
plt.show()
```

Transformaciones y procesamiento

```python
#checar la estacionalidad de mis series, prueba de dickey fuller
def test_stationarity(series, series_name):
    result = adfuller(series, autolag='AIC')
    
    print(f'\nDickey-Fuller Test para la serie {series_name}:')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value}')
    
    # resultado de la prueba
    if result[1] <= 0.05:
        print("\nConclucion: Estacionaria (reject H0)")
    else:
        print("\nConclucion: No estacionaria (fail to reject H0)")
    print("-"*70)



# Analizar cada serie
for series_id in mine:
    serie = df1[df1['V1'] == series_id].iloc[:, 1:].dropna(axis=1).values.flatten()
    test_stationarity(serie, series_id)
```

Con esta prueba de Dickey/Fuller, se pudo determinar lo siguiente:

Podria ser conveniente ahora descomponer la serie en periodos de 24 ya que son datos por hora. Esto me ayudaria a ver la tendencia de cada serie, estacionalidad, y los resicuales.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

def plot_decomposition(series, series_name, period=24):  #periodos de 24 ya que son datos por hora
    """
    Perform and plot seasonal decomposition of the time series
    """

    decomposition = seasonal_decompose(series, model='additive', period=period)
    
    # para ver mis series
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15, 12))
    
    # serie original
    ax1.plot(decomposition.observed)
    ax1.set_title(f'Original Series - {series_name}')
    ax1.grid(True)
    
    #tendencia
    ax2.plot(decomposition.trend)
    ax2.set_title('Trend')
    ax2.grid(True)
    
    # estacionalidad
    ax3.plot(decomposition.seasonal)
    ax3.set_title('Seasonal')
    ax3.grid(True)
    
    # residuales
    ax4.plot(decomposition.resid)
    ax4.set_title('Residual')
    ax4.grid(True)
    
    plt.tight_layout()
    plt.show()


#ver cada serie
for series_id in mine   :
    print(f"\nDecomposing series {series_id}")
    serie = df1[df1['V1'] == series_id].iloc[:, 1:].dropna(axis=1).values.flatten()
    
    try:
        plot_decomposition(serie, series_id)
    except Exception as e:
        print(f"Error decomposing series {series_id}: {str(e)}")
```
